Задача 1. Обработка пользовательского ввода

```python
try:
    # Запрашиваем ввод чисел
    num1 = float(input("Entered 1st number: "))
    num2 = float(input("Entered 2nd number: "))

    # Пытаемся выполнить деление
    result = num1 / num2
    print("Divide result:", result)

# Обработка деления на ноль
except ZeroDivisionError:
    print("Error: division by zero is not possible!")

# Обработка некорректного ввода данных
except ValueError:
    print("Error: enter numeric values!")

# Обработка других непредвиденных ошибок
except Exception as ex:
    print("An unexpected error has occurred:", ex)
```

Задача 2. Создание пользовательских исключений

```python
import math

# Создаем собственное исключение
class NegativeNumberError(Exception):
    # Исключение вызывается, если передано отрицательное число.
    pass


def sqrt_number(num):
    # Функция возвращает квадратный корень числа.
    # Вызывает NegativeNumberError, если число отрицательное
    if num < 0:
        raise NegativeNumberError("It is impossible to calculate the square root of a negative number!")
    return math.sqrt(num)


# Основная часть программы
try:
    number = float(input("Entered number: "))
    result = sqrt_number(number)
    print(f"The square root of {number} equal to {result}")

except NegativeNumberError as ex:
    print("Error:", ex)

except ValueError:
    print("Error: you need to enter a numeric value.!")

except Exception as ex:
    print("An unexpected error has occurred:", ex)
```

Задача 3. Обработка ошибок файлового ввода-вывода

```python
# Программа для чтения содержимого файла с обработкой ошибок

try:
    # Запрашиваем у пользователя имя файла
    filename = input("Enter the file name to read: ")

    # Пытаемся открыть файл и прочитать его содержимое
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()
        print("\nFile Contents:\n")
        print(content)

# Ошибка: файл не найден
except FileNotFoundError:
    print("Error: the file was not found. Check the file name and try again..")

# Ошибка: нет доступа к файлу
except PermissionError:
    print("Error: insufficient permissions to read the file.")

# Ошибка: другие непредвиденные проблемы
except Exception as ex:
    print("An error occurred while working with the file:", ex)
```
